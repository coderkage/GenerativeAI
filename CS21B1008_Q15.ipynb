```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**4 - 3*x**3 + 2

def grad_f(x):
    return 4*x**3 - 9*x**2

def gradient_descent(eta, x_init, steps):
    x_values = [x_init]
    for _ in range(steps):
        x_new = x_values[-1] - eta * grad_f(x_values[-1])
        x_values.append(x_new)
    return x_values
```

```python
learning_rates = [0.1]

x_init = 2.0
steps = 50

results = {}
for eta in learning_rates:
    results[eta] = gradient_descent(eta, x_init, steps)

plt.figure(figsize=(15, 5))
x_range = np.linspace(2.0, 2.5, 400) 
y_range = f(x_range)

for i, eta in enumerate(learning_rates):
    plt.subplot(1, 3, i + 1)
    plt.plot(x_range, y_range, label='f(x)')
    x_vals = results[eta]
    y_vals = [f(x) for x in x_vals]
    plt.plot(x_vals, y_vals, 'o-', label=f' = {eta}')
    plt.title(f'Gradient Descent (={eta}), steps = {steps}')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.grid(True)

plt.tight_layout()
plt.show()
```
